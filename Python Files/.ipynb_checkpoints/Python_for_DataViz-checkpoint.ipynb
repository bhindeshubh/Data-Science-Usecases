# 14/08/2024

```python
import pandas as pd # Data Manipulation
import numpy as np # Scientific Calculations
import matplotlib.pyplot as plt # 2D Plotting library
import seaborn as sns # Data Visualization library
```

```python
data = sns.load_dataset('penguins').dropna()
data.head()
```

```python
data['island'].value_counts().plot(kind='bar')
```

```python
data.groupby(['island','species']).count()[['bill_length_mm']].unstack()
```

### Stacked Bar Chart

```python
data.groupby(['island','species']).count()[['bill_length_mm']].unstack().plot(kind='bar',stacked=True,table=True)
plt.xlabel('')
plt.xticks([])
plt.show()
```

```python
data.groupby(['species','sex']).count()[['bill_depth_mm']].unstack().plot(kind='pie',subplots=True)
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.tight_layout()
```

```python
data.groupby(['species','sex']).count()[['bill_depth_mm']].unstack().plot(kind='bar',stacked=True,table=True)
plt.xlabel('')
plt.xticks([])
plt.show()
```

### Donut Pie Chart

```python
value = lambda x: (x.value_counts().keys(),x.value_counts().values)
x,values = value(data.species)
```

```python
plt.pie(values,labels=x,autopct="%1.1f%%",wedgeprops=dict(width=0.5))
plt.show()
```

```python
x,y = data.bill_length_mm, data.bill_depth_mm

plt.scatter(x,y,alpha=0.5)
plt.xlabel(x.name)
plt.ylabel(y.name)
plt.legend(pd.DataFrame(x).corrwith(y))
plt.show()
```

```python
data.dtypes
```

```python
c = 0
for i in data.columns:
    if data[i].dtypes == 'float64':
        plt.subplot(3,4,c+1)
        plt.scatter(x=data[i].value_counts().keys(),y=data[i].value_counts().values,alpha=0.5)
        plt.show()
    c+=1
```
