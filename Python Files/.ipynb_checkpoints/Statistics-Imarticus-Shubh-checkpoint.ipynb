```python
X, Y = df1.drop('MedHouseVal',axis=1),df1.MedHouseVal
```

```python
plt.axhline(0.10, linestyle='--', color='red', label='Strong')
plt.axhline(-0.10, linestyle='--', color='green', label='Weak')
X.corrwith(Y).plot(kind='bar', label='Value')
plt.title('Correlation Plot')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.legend()
```

```python
thresholds = {
    'Strong Negative': -0.5,
    'Moderate Negative': -0.3,
    'Weak Negative': -0.1,
    'Weak Positive': 0.1,
    'Moderate Positive': 0.3,
    'Strong Positive': 0.5
}

threshold_colors = {
    'Strong Negative': 'red',
    'Moderate Negative': 'blue',
    'Weak Negative': 'green',
    'Weak Positive': 'orange',
    'Moderate Positive': 'purple',
    'Strong Positive': 'brown'
}

corr = X.corrwith(Y)
num_bars = len(X.columns)
bar_colors = plt.cm.viridis(np.linspace(0, 1, num_bars))

plt.figure(figsize=(15, 12))
bars = plt.bar(corr.keys(), height=corr.values, color=bar_colors)

plt.xticks([],fontsize=7, rotation=45, c='red')

for label, threshold in thresholds.items():
    color = threshold_colors[label]
    plt.axhline(threshold, color=color, linestyle='--', linewidth=2, label=label)

plt.legend()
plt.show()
```

## Probability

```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Loan_data.csv', index_col=0)

df.head()
```

```python
df.mode()
```

```python
a = df.Gender.fillna('Male')
b = df.Married.fillna('Yes')
c = df.Dependents.fillna(0)
d = df.Education.fillna('Graduate')
e = df.Self_Employed.fillna('Yes')
f = df.ApplicantIncome.fillna(2500)
g = df.CoapplicantIncome.fillna(0.0)
h = df.LoanAmount.fillna(120.0)
i = df.Loan_Amount_Term.fillna(360.0)
j = df.Credit_History.fillna(1.0)
k = df.Property_Area.fillna('Semiurban')
l = df.Loan_Status.fillna('Y')
```

```python
df1 = pd.DataFrame((a,b,c,d,e,f,g,h,i,j,k,l)).transpose()
df1.head()
```

```python
df1.Gender.isnull().sum()
```

```python
df2 = df.select_dtypes(object)
df2.head()
```

```python
df3 = df2.drop('Loan_Status', axis=1)
df3.head()
```

```python
# Contingency table
# Probability of doing loan defaulting based on Gender
cont_table = pd.crosstab(df1.Loan_Status,df1.Gender,margins=True)
cont_table/len(df1)
```

```python
# Contingency table
# Probability of doing loan defaulting based on Marriage Status
cont_table = pd.crosstab(df1.Loan_Status,df1.Married,margins=True)
cont_table/len(df1)
```

```python
pd.crosstab
```

```python
# Contingency table
# Probability of doing loan defaulting based on Education
cont_table = pd.crosstab(index=df2.Loan_Status,columns=[df3['Gender'],df3['Married'],df3['Dependents'],df3['Education'],df3['Self_Employed'],df3['Property_Area'],df2['Loan_Status']], margins=True)
cont_table/len(df2)
```

```python
data_cleaned1 = data.copy()



for columns in num_df.columns:
    print(columns)
    try: 
        bins = list(num_df[columns].quantile([.25,.5]).values)
        bins.append(num_df[columns].max())
        bins.insert(0,(num_df[columns].min()-1))
        labels = ['low','high','Extreme']
        data_cleaned1[columns] = pd.cut(num_df[columns],labels=labels,bins=bins)
    except:
        print(f'I did not process: {columns}')
```
