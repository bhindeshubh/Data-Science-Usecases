# **Data Preprocessing**

```python
# No. of rows and columns
print(f'No. of Rows: {df.shape[0]} \nNo. of Columns: {df.shape[1]}')
```

```python
# Seperating float values into a new dataframe
df_float = df.select_dtypes(float)
print(f'No. of Rows: {df_float.shape[0]} & No. of Columns: {df_float.shape[1]}')
df_float.head()
```

```python
# Seperating integer values into a new dataframe
df_int = df.select_dtypes(int)
print(f'No. of Rows: {df_int.shape[0]} & No. of Columns: {df_int.shape[1]}')
df_int.head()
```

```python
# Seperating object values into a new dataframe
df_obj = df.select_dtypes(object)
print(f'No. of Rows: {df_obj.shape[0]} & No. of Columns: {df_obj.shape[1]}')
df_obj.head()
```

```python
# The month of highest arrival
df_obj['arrival_date_month'].mode()
```

```python
# No. of hotels in dataset
df_obj['hotel'].unique()
```

```python
# Filtering the data based on hotel names
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel
```

```python
# Filtering the data based on hotel names
city_hotel = df[df['hotel']=='City Hotel']
city_hotel
```

```python
# Hotelwise cancellations

h1 = df[df['hotel'] == df['hotel'].value_counts().keys()[0]]
v1 = round(h1['is_canceled'].sum()/len(h1['is_canceled']),2)

h2 = df[df['hotel'] == df['hotel'].value_counts().keys()[0]]
v2 = round(h2['is_canceled'].sum()/len(h2['is_canceled']),2)

if v1 > v2:
    print('Resort Hotel has more cancellations')
else:
    print('City Hotel has more cancellations')
```

```python
# Above thing using for loop
hotels = {}

for i in df['hotel'].value_counts().keys():
    print(i)
    hotel1 = df[df['hotel'] == i]
    hotels[i] = round(hotel1['is_canceled'].sum()/len(hotel1['is_canceled']),2)
print(hotels)    

if list(hotels.values())[0] > list(hotels.values())[1]:
    print(f'{list(hotels.keys())[0]} has higher cancellations')
else:
    print(f'{list(hotels.keys())[1]} has higher cancellations')
```

```python
# Hotels cancellations for each year for city hotel
hotels1 = {}

for i in city_hotel['arrival_date_year'].value_counts().keys():
    print(i)
    hotel1 = city_hotel[city_hotel['arrival_date_year'] == i]
    hotels1[i] = round(hotel1['is_canceled'].sum()/len(hotel1['is_canceled']),2)
print(hotels1)    

if list(hotels1.values())[0] > list(hotels1.values())[1]:
    print(f'{list(hotels1.keys())[0]} has higher cancellations')
else:
    print(f'{list(hotels1.keys())[1]} has higher cancellations')
```

```python
hotels2 = {}

for i in resort_hotel['arrival_date_year'].value_counts().keys():
    print(i)
    hotel2 = resort_hotel[resort_hotel['arrival_date_year'] == i]
    hotels2[i] = round(hotel2['is_canceled'].sum()/len(hotel2['is_canceled']),2)
print(hotels2)    

if list(hotels2.values())[0] > list(hotels2.values())[1]:
    print(f'{list(hotels2.keys())[0]} has higher cancellations')
else:
    print(f'{list(hotels2.keys())[1]} has higher cancellations')
```

```python
hotel_1 = pd.DataFrame(hotels1, index=[0])
hotel_2 = pd.DataFrame(hotels2,index=[0])
hotel_1.head()
```
