1. Write a Python function that takes a tuple of numbers as input and returns a new tuple with each element squared.

```python
a = (1, 15, 24, 9, 16) # Creating a tuple of numbers

b = [] # Creating a blank list

# Loop to store squared elements of tuple inside the blank list
for i in a:
    b.append(i**2)

c = tuple(b) # Creating the tuple of list elements

print(c)
```

2. Write a Python function that removes all occurrences of a specified value from a list.

```python
l = [16, 'Hello', 16, 25, "World", 27] # Creating a list

# A loop to remove all occurences of 16 inside the list
for i in l:
    if i == 16:
        l.remove(i)

# Printing the list
print(l)
```

3. Write a Python function that takes a number as input and returns whether it is positive, negative, or zero using an if-elif-else structure.

```python
a = int(input("Enter a Number:")) # Taking input from user

if a == 0:
    print("The number is Zero")   # Condition to return Zero
elif a > 0:
    print("The number is Positive") # Condition to return Positive
elif a < 0:
    print("The number is Negative") # Condition to return Negative
```

5. Create a Pandas DataFrame from a dictionary with the following data:
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
add a new column called ""Score"" with the values [85, 90, 88, 76]. Finally, display the DataFrame.


```python
# Importing pandas library
import pandas as pd 

# Defining data needed inside the dataframe
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df1 = pd.DataFrame(data) # Creating dataframe of the data

df2 = pd.DataFrame({"Score": [85, 90, 88, 76]}) # Creating a dataframe of column to be added inside the original dataframe

df1.join(df2) # Adding the column in the dataframe

df1
```

7. Task: Using Matplotlib, create a line plot for the following data:
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
Label the x-axis as ""X Values"" and the y-axis as ""Y Values"". Add a title ""Simple Line Plot"" and display the plot.

```python
# Importing Matplotlib library
import matplotlib.pyplot as plt

# Defining values to be plotted
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x,y)                   # Command to plot line graph
plt.title("Simple Line Plot")   # Command to add title
plt.xlabel("X-Values")          # Command to label the X-axis
plt.ylabel("Y-values")          # Command to label the Y-axis
plt.show()                      # Command to display the final plot
```

9. Task: Given the following data:
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
Create atleast 3 discrete plot. Label the charts with their respective categories.

```python
# Data to be plotted
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
```

```python
sns.barplot(x=categories,y=values) # Discrete Plot 1(Bar Graph)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Discrete Plot 1")
plt.show()
```

```python
plt.pie(x=values,labels=categories) # Discrete plot 2(Pie Plot)
plt.title("Discrete Plot 2") # Giving title to the plot
plt.show() # Displaying the plot
```

```python
plt.barh(categories,values)
plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Discrete Plot 3")
plt.show()
```
